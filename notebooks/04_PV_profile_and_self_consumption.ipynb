```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.io import loadmat

load_mat = loadmat("../data/raw/load_yearly.mat")
pv_mat = loadmat("../data/raw/PV_production_yearly.mat")
price_mat = loadmat("../data/raw/prices_yearly.mat")
```

```python
print(load_mat.keys())
print(pv_mat.keys())
print(price_mat.keys())
```

```python
load_year = load_mat["load_all_year"].flatten()
pv_factor = pv_mat["zeta_PV"].flatten()
price_year = price_mat["prices_all_year"].flatten()

print("Load shape :", load_year.shape)
print("PV shape   :", pv_factor.shape)
print("Price shape:", price_year.shape)
```

```python
print("Load")
print("Min :", load_year.min())
print("Max :", load_year.max())
print("Mean:", load_year.mean())

print("\nPV")
print("Min :", pv_factor.min())
print("Max :", pv_factor.max())
print("Mean:", pv_factor.mean())

print("\nPrices")
print("Min :", price_year.min())
print("Max :", price_year.max())
print("Mean:", price_year.mean())
```

```python
plt.figure(figsize=(12,4))
plt.plot(load_year)

plt.title("Yearly Building Load")
plt.xlabel("Hour of Year")
plt.ylabel("Power [kW]")
plt.grid()

plt.show()
```

PV profile

```python
plt.figure(figsize=(12,4))
plt.plot(pv_factor)

plt.title("Yearly PV Production Factor")
plt.xlabel("Hour of Year")
plt.ylabel("Normalized PV Output")
plt.grid()

plt.show()
```

Electricity prices

```python
plt.figure(figsize=(12,4))
plt.plot(price_year)

plt.title("Yearly Electricity Price")
plt.xlabel("Hour of Year")
plt.ylabel("Price")
plt.grid()

plt.show()
```
